## 1. Concatenation: `concat()` function 

Concatenation adds together DataFrames: dimensions should match along the axis we are concatenating on. 

We can use **pd.concat** and pass in a list of DataFrames to concatenate together:

### Example 1

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']}, index=[0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']}, index=[4, 5, 6, 7]) 
df2
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']}, index=[8, 9, 10, 11])
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3]).drop_duplicates().reset_index(drop=True)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

## 2. Merging: `merge()` function

The `merge()` function allows you to merge DataFrames together using a similar logic as merging SQL Tables together.

### Example 2

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
left 
```

```python
right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3'], 'key': ['K0', 'K1', 'K2', 'K3'],})   
right
```

```python
pd.merge(left,right,how='left',on='key')
```

```python
pd.merge(left,right,how='left',left_on='key', right_on='key',indicator = 'key_merge')  # if the join key names are not the same
```

```python
pd.merge(left,right,how='inner',on='key')  # inner by default
```

Or to show a more complicated example:

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
right
```

```python
pd.merge(left, right, on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='inner', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
left.merge(right, how='left', on =['key1','key2'])
```

## 3. Joining: join() function
Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame.

```python
left1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                      'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 
left1
```

```python
right2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                       'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
right2
```

```python
left1.join(right2)
```

```python
left1.join(right2, how='right')
```

```python
left1.join(right2, how='outer')
```

## 4. Append New Row to Your Dataframe

```python
df1 = pd.DataFrame({'letter':['A','B','C','D'],
                    'number':[1,2,3,4]})
df2 = pd.DataFrame({'letter':['C','D','E','F'],
                    'number':[3,4,5,6]})
```

```python
# Union with pd.concat()
df3 = pd.concat([df1,df2]).drop_duplicates().reset_index(drop=True)
df3
```

```python
# Concatenate dataframes horizontally
df4 = pd.concat([df1,df2], axis=1)
df4
```

```python
new_row = pd.Series(['Z',26],index=df3.columns)
df3.append(new_row,ignore_index=True)
```
