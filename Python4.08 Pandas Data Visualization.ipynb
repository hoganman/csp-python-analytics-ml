## 2. Plot Styles

Matplotlib has [style sheets](http://matplotlib.org/gallery.html#style_sheets) you can use to make your plots look a little nicer. These style sheets include plot_bmh, plot_fivethirtyeight, plot_ggplot and more. They basically create a set of style rules that your plots follow. I recommend using them, they make all your plots have the same look and feel more professional. You can even create your own if you want your company's plots to all have the same look (it is a bit tedious to create on though).

Here is how to use them.

**Before plt.style.use() your plots look like this:**

```python
df1['A'].hist()
```

```python
df1['A'].hist(bins=100)  # add matplotlib arguments
```

Call the style:

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Now your plots look like this:

```python
df1['A'].hist()
```

```python
df1['A'].plot.hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

### 5) Scatter Plot

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

You can use c to color based off another column value
Use cmap to indicate colormap to use. 
For all the colormaps, check out: http://matplotlib.org/users/colormaps.html

Or use s to indicate size based off another column. s parameter needs to be an array, not just the name of a column:

```python
df1.dtypes
```

```python
df1['A'].value_counts
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*50)
```

### 7) Hexagonal Bin Plot

Useful for Bivariate Data, alternative to scatterplot:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=35)
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```

### Example

```python
iris_df = pd.read_excel('iris.xlsx')
iris_df.head()
```

```python
#plt.style.use('dark_background')
plt.style.use('bmh')
iris_df.hist()
```

```python
plt.style.use('dark_background')
iris_df[['Sepal_Length','Sepal_Width']].plot.hist(alpha=.8)
```

```python
iris_df.dtypes
```

```python
iris_df['Species'].value_counts()
```

```python
plt.style.use('bmh')
colors = {'Iris-versicolor':'red','Iris-setosa':'green','Iris-virginica':'blue'}
iris_df['colors'] = iris_df['Species'].map(colors)
iris_df.plot.scatter(x = 'Sepal_Width',y = 'Sepal_Length', color = iris_df['colors']);
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris_df, figsize = (16,8))
```

Hopefully you can see why this method of plotting will be a lot easier to use than full-on matplotlib, it balances ease of use with control over the figure. A lot of the plot calls also accept additional arguments of their parent matplotlib plt. call.
