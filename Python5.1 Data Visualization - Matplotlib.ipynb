## 1. Introduction to Matplotlib

Matplotlib is `the major and most popular plotting library of data visualization with Python`. It was created by John Hunter. He created it to try to replicate **MatLab's** (another programming language) plotting capabilities in Python. So if you happen to be familiar with MatLab, Matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major benefits of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* It was designed to have a similar feel to MatLab's graphical plotting

#### Matplotlib allows you to create reproducible figures programmatically.


### 2) The Object Oriented Method

The main idea in using the more formal Object Oriented method is to instanitate/create figure objects and then call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it and have more controls comparing to functional method.

To begin we create a figure instance. Then we can add axes to that figure:

````python
fig = plt.figure()  # create a figure object: a magical blank canvas
````

````python
fig = plt.figure()  # blank canvas
axes = fig.add_axes([0.1, 0.3, 0.4, 0.5])    #  of fig object, call the add_axes method
                                           #  10% - left of axis
                                           #  20% - bottom of axis
                                           #  80% - width of axis
                                           #  70% - height of axis
                                           #  A 4-length sequence of [left, bottom, width, height] quantities.
````

````python
# Exercise
fig = plt.figure()  # blank canvas
axes = fig.add_axes([0.5, 0.2, 0.4, 0.7])
````

````python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.2, 0, 0.8, 0.8]) # left, bottom, width, height (0-1: the percent of the canvas taking) - width, height in inches(?)

# Plot on that set of axes
axes.plot(x, y,'r')    # r for red
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
````

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure:

````python
# Creates blank canvas
fig = plt.figure()

axes = fig.add_axes([0.1, 0.5, 0.9, 0.8]) # main axes: 10% from the left, 10% from the bottom, 80% width, 80% height
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
````

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

````python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots() # can specify number of rows and columns; fig = plt.figure() & axes1 = fig.add_axes

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
````

````python
# Empty canvas of 1 by 3 subplots
fig, axes = plt.subplots(nrows=1, ncols=3)
plt.tight_layout()
````

````python
axes # is a list of matplotlib axes which was manually created using fig.add_axes()
````

````python
# Empty canvas of 1 by 4 subplots
fig, axes = plt.subplots(nrows=1, ncols=4)  # add axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) automatically based on rows and cols

````

### 1) we can iterate through the axes object:

````python
# Empty canvas of 1 by 4 subplots
fig, axes = plt.subplots(nrows=1, ncols=5)  # add axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) automatically based on rows and cols

# we can iterate through the axes object:
for curr_ax in axes:
    curr_ax.plot(x,y)
plt.tight_layout()
````

````python
axes
````

Then you can specify the number of rows and columns when creating the subplots() object:

### 2) We can also index the axes object:

````python
# Empty canvas of 1 by 3 subplots
fig, axes = plt.subplots(nrows=1, ncols=3)

#plt.tight_layout()

# We can also index the axes object
axes[0].plot(x,y,'r')
axes[0].set_title('1st Graph')

axes[1].plot(y,x,'g')
axes[1].set_title('2nd Graph')

axes[2].plot(y,x,'b')
axes[2].set_title('3rd Graph')

plt.tight_layout()  # recommend to always put this in the end of commends
````

````python
# Axes is an array of axes to plot on
axes
````

A common issue with matplolib is overlapping subplots or figures. We can use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

````python
fig, axes = plt.subplots(nrows=1, ncols=4)

for ax in axes:
    ax.plot(x, y, 'g--')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig    
plt.tight_layout()
````

### Figure Size, Aspect Ratio and DPI

**Matplotlib** allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

````python
fig = plt.figure(figsize=(6,2), dpi=100) # or default dpi
````

````python
fig = plt.figure(figsize=(6,2), dpi=100) # or default dpi
ax = fig.add_axes([0.1,0.1,0.6,0.6])
ax
````

````python
ax.plot(x,y)
````

````python
fig = plt.figure(figsize=(8,4), dpi=80) # width and height in inches dpi=dot/ppixel per inch
ax = fig.add_axes([0.1,0.1,0.6,0.8])
ax.plot(x,y)
````

The same arguments can also be passed to layout managers, such as the `subplots` function:

````python
fig, axes = plt.subplots(figsize=(8,3))  # 8 inches by 4 inches

axes.plot(x, y, 'r')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('plot-title');

plt.tight_layout()
````

### Legends, Labels and Titles

Now that we have covered the basics of how to create a figure canvas and add axes instances to the canvas, let's look at how to add a figure with titles, axis labels, and legends.

**Figure titles**

A title can be added to each axis instance in a figure. To set the title, use the `set_title()` method in the axes instance:

````python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title("customized title");
````

**Axis labels**

Similarly, with the methods `set_xlabel` and `set_ylabel`, we can set the labels of the X and Y axes:

````python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y)

ax.set_xlabel("xlable")
ax.set_ylabel("ylabel")
ax.set_title("charttitle")
````

**Legends**

You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure: 

````python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
#ax.plot(x, x**2)             
#ax.plot(x, x**3)

ax.plot(x, x**2, label="x**2")             # Step 1: Add labels for ax.legend to referring
ax.plot(x, x**3, label="x**3")

ax.legend()                                # Step 2: Add legend()
````

Notice how are legend overlaps some of the actual plot!

The **legend** function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be drawn. The allowed values of **loc** are numerical codes for the various places the legend can be drawn. See the [Location documentation page](https://matplotlib.org/stable/) for details. Some of the most common **loc** values are:

````python
# Lots of options....

ax.legend(loc=0) # best
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner
...
ax.legend(loc=10) # Center 

# .. many more options are available

# Most common to choose
ax.legend(loc=0) # let matplotlib decide the best/optimal location
fig
````

````python
ax.legend(loc=(0.2,0.8))  # Alternatively, use 2-tuple giving x,y of lower-left corner of the legend in axes coordinates
fig
````

### Setting colors, linewidths, linestyles

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linestyles. There is the basic MATLAB like syntax:

#### Colors with MatLab like syntax

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

````python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='r',lw=2,ls='--') # or linestyle
````

````python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',lw=2,ls='--')   # lw=linewidth  ls=linestyle
````

````python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='--', alpha=1)   # lw=linewidth  ls=linestyle
````

````python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line

# Add xlim and ylim
ax.set_xlim([0,3])
ax.set_ylim([0,10])
````

#### Colors with the `color=` parameter

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

````python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="green", lw=2, ls='-', alpha=0.6, 
         marker='o', markersize=10, markerfacecolor='red',
         markeredgewidth=3, markeredgecolor='green')  # alpha --> half-transparant
ax.plot(x, x+2, color="#8B008B")           # RGB hex code for customed colors
ax.plot(x, x+3, color="#FFCCE9")           # RGB hex code for customed colors
````

### Control over Axis Appearance

In this section we will look at controlling axis sizing properties in a matplotlib figure.

### Plot Range

We can configure the ranges of the axes using the **set_ylim** and **set_xlim** methods in the axis object, or **axis('tight')** for automatically getting "tightly fitted" axes ranges:

````python
fig, axes = plt.subplots(1, 3, figsize=(12, 2))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
````
